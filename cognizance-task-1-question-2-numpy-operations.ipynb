```python
import numpy as np 
arr = [3,4,-2,5,1,-5,25,-6,9]
arr1= [4,16,64,25]

arr = np.array([3, 4, -2, 5, 1, -5, 25, -6, 9])

arr_reshaped = np.reshape(np.sort(arr1), (2, 2))
print("Array arranged and reshaped:")
print(arr_reshaped)

print("\nUNARY FUNCTIONS:\n")

print("ABSOLUTE VALUE\n")
abs_arr1 = np.abs(arr)
abs_arr2 = np.abs(arr1)
print("Absolute value of each element:")
print(abs_arr1,"\n")
print(abs_arr2,"\n")

print("SQUARE ROOT")
sqrt_arr2= np.sqrt(arr1)
print("\nSquare root of each element:")
print("Square root of negative numbers does not exist\n")
print(sqrt_arr2,"\n")

print("EXPONENTIAL VALUE")
exp_arr1 = np.exp(arr)
exp_arr2 = np.exp(arr1)
print("\nExponential value of each element:")
print(exp_arr1,"\n")
print(exp_arr2,"\n")

print("SQUARE")
square_arr1 = np.square(arr)
square_arr2 = np.square(arr1)
print("\nSquare of each element:")
print(square_arr1,"\n")
print(square_arr2,"\n")

print("SINE VALUE")
sin_arr1 = np.sin(arr)
sin_arr2 = np.sin(arr1)
print("\nSine value of each element:")
print(sin_arr1,"\n")
print(sin_arr2,"\n")

print("COSINE VALUE")
cos_arr1 = np.cos(arr)
cos_arr2 = np.cos(arr1)
print("\nCosine value of each element:")
print(cos_arr1,"\n")
print(cos_arr2,"\n")

print("TANGENT VALUE")
tan_arr1 = np.tan(arr)
tan_arr2 = np.tan(arr1)
print("\nTangent value of each element:")
print(tan_arr1,"\n")
print(tan_arr2,"\n")

weight_players = np.array([75, 68, 82, 87, 69])
height_players = np.array([170, 165, 175, 180, 168])
print("\n\nWeight of players:",weight_players,"\n")
print("Height of players:",height_players,"\n")

print("MEAN OF WEIGHT OF THE PLAYERS\n")
mean_weight = np.mean(weight_players)
print("Mean weight of the players:", mean_weight,"\n")

print("SUM OF THE HEIGHT OF THE PLAYERS\n")
sum_height = np.sum(height_players)
print("Sum of height of the players:", sum_height,"\n")

print("STANDARD DEVIATION OF THE WEIGHT OF THE PLAYERS\n")
std_weight = np.std(weight_players)
print("Standard deviation of weight of the players:", std_weight,"\n")

print("STANDARD DEVIATION OF THE HEIGHT OF THE PLAYERS\n")
std_height = np.std(height_players)
print("Standard deviation of height of the players:", std_height,"\n")


```

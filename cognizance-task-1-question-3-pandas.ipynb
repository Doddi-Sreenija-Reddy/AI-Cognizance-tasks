```python
import pandas as pd
import numpy as np
l = [10,20,30,40,50]
d1 = {"a": 10, "b": 20, "c": 30, "d": 40, "e": 50}
print("List l:",l,"\n")
print("Dictionary d1:",d1,"\n")
print("CONVERSION OF LIST INTO SERIES\n")
series_list = pd.Series(l)
print("Series from list l:")
print(series_list,"\n")

print("CONVERT DICTIONARY TO A SERIES\n")
series_dictionary = pd.Series(d1)
print("Series from dictionary d1:")
print(series_dictionary,"\n")

print("ADDITION OF TWO SERIES\n")
print("Series 1:\nIndia 1\nSrilanka 2\nBangladesh 3\nRussia 4\nEngland 5\n")
print("Series 2:\nIndia 6\nSrilanka 7\nBangladesh 8\nRussia 9\nEngland 10\n")
series1 = pd.Series([1, 2, 3, 4, 5], index=['India', 'Srilanka', 'Bangladesh', 'Russia', 'England'])
series2 = pd.Series([6, 7, 8, 9, 10], index=['India', 'Srilanka', 'Bangladesh', 'Russia', 'England'])
added_series = series1 + series2
print("Added Series:")
print(added_series,"\n")

print("CREATION OF DATA FRAME\n")
np.random.seed(0)
data = np.random.randint(0, 100, (5, 4))
data_frame = pd.DataFrame(data, columns=['W', 'X', 'Y', 'Z'], index=['A', 'B', 'C', 'D', 'E'])
print("DataFrame:")
print(data_frame,"\n")

print("DATAFRAME INTO EXCEL FILE\n")
data_frame.to_excel('dataframe.xlsx', index=True)
print('/kaggle/working/dataframe.xlsx')
print("DataFrame written to 'dataframe.xlsx' file.\n")

print("ACCESS A COLUMN AND A ROW OF THE DATAFRAME\n")
column_X = data_frame['X']
row_A = data_frame.loc['A']
print("Accessed Column X:")
print(column_X)
print("\nAccessed Row A:")
print(row_A,"\n")

print("ACCESS DATA WITH INDEX VALUES OF ROW A\n")
data_loc_A = data_frame.loc['A']
data_iloc_A = data_frame.iloc[0]
print("Data using loc for row A:")
print(data_loc_A)
print("\nData using iloc for row A:")
print(data_iloc_A,"\n")

print("SELECTION OF A SUBSET OF DATA\n")
subset_loc = data_frame.loc[['A', 'E'], ['W', 'Z']]
print("Subset of data using loc for rows A and E, and columns W and Z:")
print(subset_loc,"\n")

print("DROP A ROW")
df_dropped = data_frame.drop('A')
print("\nDataFrame after dropping the first row:")
print(df_dropped,"\n")
```
